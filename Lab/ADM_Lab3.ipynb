```python
def dateparse(time_in_secs):
    return pd.to_datetime(time_in_secs, unit='s')
```

```python
dataset = pd.read_csv('steam_reviews.csv', header='infer',
                     nrows=2000000,
                     parse_dates=['timestamp_created',
                                 'timestamp_updated', 'author.last_played'],
                                 date_parser=dateparse)
```

```python
import matplotlib.pyplot as plt
```

```python
dataset['language'].value_counts().plot()
```

```python
dataset['language'].value_counts().plot.bar()
```

```python
dataset['language'].value_counts().plot.bar(figsize = (18, 7), xlabel = 'langauge')
```

```python
plt.rcParams['figure.figsize'] = (8, 8)
plt.pie(dataset['language'].value_counts(),
        labels = dataset.language.unique(),
        explode = [0.1 for value in range(0, dataset.language.nunique())],
        shadow = True, autopct = '%.1f%%')
plt.title('Languages', fontsize = 20)
plt.axis('off')
plt.show()
```
